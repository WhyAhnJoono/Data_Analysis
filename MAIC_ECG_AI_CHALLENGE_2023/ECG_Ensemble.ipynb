```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
submission = pd.read_csv('./dataset/submission.csv')
AutoML_AJH_ver2 = pd.read_csv('./submission/AutoML_AJH_ver2.csv')
AutoML_AJH_ver3 = pd.read_csv('./submission/AutoML_AJH_ver3.csv')
Autogluon_KYJ_integrated = pd.read_csv('./submission/통합_Autogluon_KYJ_230921.csv')
Autogluon_KYJ_separated = pd.read_csv('./submission/개별_Autogluon_KYJ_230922.csv')
```

```python
ensemble_result = (AutoML_AJH_ver2['AGE'] + AutoML_AJH_ver3['AGE'] + Autogluon_KYJ_integrated['AGE'] + Autogluon_KYJ_integrated['AGE']) / 4
ensemble_result
```

```python
# child, adult 예측 결과 분리
child_result = ensemble_result[0:1100]
adult_result = ensemble_result[1100:]
```

```python
# child 예측값 - 개월수에 맞게 치환
child_age_list = np.arange(1/12, 19, 1/12)

def find_closest_value(input_value, value_list):
    closest_value = None
    min_difference = float('inf')

    for value in value_list:
        difference = abs(input_value - value)
        if difference < min_difference:
            min_difference = difference
            closest_value = value

    return closest_value

child_age_value = []
for val in child_result:
    child_age_value.append(find_closest_value(val, child_age_list))
    
child_age_value
```

```python
# child 예측값 + adult 예측값
pred_list = []
[pred_list.append(val) for val in child_age_value]
[pred_list.append(round(val)) for val in adult_result] # adult의 경우 예측값이 정수
```

```python
submission['AGE'] = pred_list
submission
```

```python
submission.to_csv('./submission/ECG_Ensemble_AJH+KYJ.csv', index=False)
```
